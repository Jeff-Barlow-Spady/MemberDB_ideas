```python
import streamlit as st
import pandas as pd 
import sqlite3
```

```python
# Database functions

def get_db_connection():
    conn = sqlite3.connect('data.db')
    return conn

def close_db_connection(conn):
    conn.close()

def create_table():
    conn = get_db_connection()
    cursor = conn.cursor()
    
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS entries (
            id INTEGER PRIMARY KEY,
            timestamp TEXT, 
            email TEXT,
            first_name TEXT, 
            last_name TEXT,
            linkedin_profile TEXT, 
            portfolio_link TEXT,
            experience_level TEXT,
            skills TEXT,
            resume_link TEXT,
            open_to_jobs INTEGER
        )
    ''')
    
    conn.commit()
    close_db_connection(conn)

# Input functions 

def get_user_input():
    timestamp = st.text_input('Timestamp') 
    email = st.text_input('Email')
    first_name = st.text_input('First Name')
    last_name = st.text_input('Last Name')
    linkedin = st.text_input('LinkedIn Profile')
    portfolio = st.text_input('Portfolio Link')
    experience = st.text_input('Experience Level')
    skills = st.text_input('Skills (comma separated)') 
    resume = st.text_input('Resume Link')
    open_to_jobs = st.checkbox('Open to Jobs?')
    
    data = [timestamp, email, first_name, last_name, linkedin, 
            portfolio, experience, skills, resume, open_to_jobs]
            
    return data

# Save data function
        
def save_data(data):
    conn = get_db_connection()
    cursor = conn.cursor()
    
    sql = '''INSERT INTO entries VALUES (?,?,?,?,?,?,?,?,?,?,?)'''   
    
    cursor.execute(sql, data)
    conn.commit()
    
    close_db_connection(conn)
    
    st.success('Data saved!')
    
# View data function

def view_data():
    conn = get_db_connection() 
    data = pd.read_sql('SELECT * FROM entries', conn)
    close_db_connection(conn)
    
    st.dataframe(data)


# Main app

def main():
    
    st.title("Data Entry App")
    
    create_table()
    
    data = get_user_input()
    save_data(data)
    
    view_data()




st
$streamlit run r'C:\Users\toast\AppData\Roaming\Python\Python310\site-packages\ipykernel_launcher.py'
```
